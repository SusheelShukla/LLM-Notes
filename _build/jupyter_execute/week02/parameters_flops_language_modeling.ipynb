## Key Concepts

### Parameters

Parameters are learnable values inside a neural network.

Examples:

* Weights
* Biases
* Embedding matrices

A trained model stores knowledge inside its parameters.

Examples:

| Model       | Approximate Parameters |
| ----------- | ---------------------- |
| BERT Base   | 110 Million            |
| BERT Large  | 340 Million            |
| GPT-2 Small | 124 Million            |
| GPT-3       | 175 Billion            |
| Llama 3 8B  | 8 Billion              |
| Llama 3 70B | 70 Billion             |

---

### Computational Complexity

Complexity measures how computation grows as input size increases.

Usually represented as:

O(n)

Examples:

| Operation             | Complexity |
| --------------------- | ---------- |
| Linear Search         | O(n)       |
| Binary Search         | O(log n)   |
| Matrix Multiplication | O(n³)      |
| Self Attention        | O(n²)      |

---

### FLOPs

FLOPs stands for:

Floating Point Operations

Examples:

* Addition
* Multiplication
* Division

Used to estimate:

* Training cost
* Inference cost
* Hardware requirements

---

Example:

Output:


The model predicts the most probable next token.

## Mathematical Foundation

### Parameters in Dense Layer

For a dense layer:

$$
Parameters = (Input \times Output) + Output
$$

where:

* Input = number of input neurons
* Output = number of output neurons
* Output term corresponds to biases

---

#### Example

Layer:

$
512 \rightarrow 1024
$

Parameters:

$
512 \times 1024 = 524288
$

Bias:

$
1024
$

Total:

$
525312
$

---

### Embedding Layer Parameters

$\boxed{\text{Parameters} = V \times d}$ Where,
- Vocabulary: $V$
- Embedding dimension: $d$
- Example:
  - Vocabulary: $50,000$
  - Embedding Dimension: $768$
  - Parameters: $50,000 \times 768 = 38,400,000$
  - Answer: $\boxed{38.4\ million\ parameters}$.

---

### Self-Attention Complexity

**Attention equation:** 

$Attention(Q,K,V) = softmax \left(\frac{QK^T}{\sqrt{d_k}}\right)V$

- Computing: $QK^T$
- requires: $O(n^2)$

where:
- n = sequence length
- This is the major computational bottleneck.

---

### FLOPs for Matrix Multiplication

- Matrix-A: $A(m \times k)$
- Matrix-B: $B(k \times n)$
- FLOPs: $2mkn$

---

#### Example

- Multiply: $1024 \times 1024$
by $1024 \times 1024$
- FLOPs: $2 \times 1024^3$
-   $2.15 \times 10^9$
- Approximately: $\boxed{2.15\ GFLOPs}$

## Detailed Explanation

### Why Parameter Count Matters

More parameters generally mean:

Advantages:

* Better representation power
* Better learning capability
* Better generalization

Disadvantages:

* More memory
* More computation
* Longer training time

---

### Why FLOPs Matter

Training cost depends heavily on FLOPs.

Examples:

| Model                  | Estimated Training FLOPs |
| ---------------------- | ------------------------ |
| GPT-2                  | ~1e20                    |
| GPT-3                  | ~3e23                    |
| Modern Frontier Models | >1e25                    |

As models scale:

* Energy consumption increases
* GPU requirements increase
* Training cost increases

---

### Transformer Complexity Problem

RNN Complexity: $O(n)$

Attention Complexity: $O(n^2)$

Example:

- Sequence length: ``` 128 tokens```
- Attention matrix: ``` 128 × 128 = 16,384 entries ```
- Sequence length: ``` 4096 tokens ```
- Attention matrix: ``` 4096 × 4096 = 16,777,216 entries```

The cost grows rapidly.

---

### Why Long Context is Hard

Modern LLMs support:

* 32K context
* 128K context
* 1M context

Attention memory grows quadratically.

This motivates:

* Flash Attention
* Sparse Attention
* Sliding Window Attention
* Linear Attention

## Introduction to Language Modeling

### What is a Language Model?

A language model estimates:

$$
P(w_1,w_2,...,w_n)
$$

Probability of a sequence of words.



### Chain Rule

$$
P(w_1,w_2,w_3) = P(w_1) P(w_2|w_1) P(w_3|w_1,w_2)
$$

Language modeling predicts the next token using previous tokens.

---

### Example

Sentence:

```text
The cat sat on the
```

Possible predictions:

| Token | Probability |
| ----- | ----------- |
| mat   | 0.62        |
| floor | 0.20        |
| chair | 0.08        |
| roof  | 0.05        |

Highest probability token is selected.

---

### Traditional Language Models

Before Transformers:

* N-gram Models
* Statistical Language Models

Example:

Bigram:

```text
I love
```

Predict:

```text
AI
```

based on frequency counts.

Problems:

* Data sparsity
* Poor generalization
* No semantic understanding

---

### Neural Language Models

Neural networks learn:

* Syntax
* Semantics
* Context

Advantages:

* Better generalization
* Context awareness
* Transfer learning

## Worked Example

### Parameter Counting

- Network:

  ```text
  Input: 784
  Hidden: 128
  Output: 10
  ```

- Bias: $128$
- Total: $100480$

- Output Layer: $128 \times 10 = 1280$

___
- Grand Total: $101770$ parameters.

### Attention Complexity Example

- Sequence: ```128 tokens```
- Attention matrix: $128^2 = 16384$
- Sequence: ```512 tokens```
- Attention matrix: $512^2 = 262144$

Increase: $16 \times$

more computation.

## Python Demonstration

### Parameter Counting

````python
def count_parameters(input_dim, output_dim):
    return input_dim * output_dim + output_dim

print(count_parameters(512,1024))
````

---

### FLOPs Estimation

````python
def matmul_flops(m,k,n):
    return 2*m*k*n

print(matmul_flops(1024,1024,1024))
````
